```python
student_name = ["Alice", "Bob", "Charlie", "Diana"]
attendance = [45, 80, 60, 30]
quiz_marks = [55, 75, 62, 90]
lms_time = [0.5, 2, 1.5, 0.8]

def assert_risk():
    for i in range(len(student_name)):
        risk_value = 0
        if attendance[i] < 50:
            risk_value += 3
        else:
            risk_value += 5

        if quiz_marks[i] < 65:
            risk_value += 2
        else:
            risk_value += 4

        if lms_time[i] < 1:
            risk_value += 1
        else:
            risk_value += 2

        if risk_value < 7:
            print(student_name[i], "Risk Student")
        else:
            print(student_name[i], "Not Risk Student")

assert_risk()
```

```python
reviews = dict({
    "Bimali" : {"M1":5, "M2":4, "M3":4},
    "Sithum" : {"M1":3, "M2":4, "M3":5},
})

ratings = dict()

for reviewer,review in reviews.items():
    for movie,rating in review.items():
        if movie not in ratings:
            ratings[movie] = []
        ratings[movie].append(rating)

def calculate_average_rating(ratings:list[int]):
    return sum(ratings)/len(ratings)

def get_highest_avg_movies():
    for mv,rating in ratings.items():
        avg_rating = calculate_average_rating(rating)
        print(f"{mv} : {avg_rating}")

get_highest_avg_movies()
```
